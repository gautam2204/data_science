```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as pylt
import seaborn as sns
```

```python
con = sqlite3.connect('Mcdonalsd.db')
df = pd.read_csv('McDonaldsMenuNutritionV3.csv',index_col=0)
cursor = con.cursor()
df.to_sql('McDonaldsMenuNutrition',con,if_exists="replace")


val = cursor.execute('''Select * From McDonaldsMenuNutrition Limit 5''')
print(val)



for i in val:
    print(i)

df.head()
```

```python
dataFrame=pd.read_sql_query(sql="Select * FROM McDonaldsMenuNutrition",con=con)
dataFrame.reset_index(inplace=True)
# dataFrame.set_index(df.columns[0],inplace=True)
dataFrame.head()
```

```python
dataFrame.describe(include="all")
```

```python
sw_plot=sns.swarmplot(x='Item',y="Sodium (mg)",data=dataFrame)
pylt.setp(sw_plot.get_xticklabels(),rotation=70)
pylt.show()
```
